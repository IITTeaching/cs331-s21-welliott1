## 1.  The `LinkedList` and `Node` classes



```python
class LinkedList:
    class Node:
        def __init__(self, val, next=None):
            self.val = val
            self.next = next

    def __init__(self):
        self.head = None
        self.count = 0

    def prepend(self, value):
        pass

    def __len__(self):
        return self.count

    def __iter__(self):
        n = self.head
        while n:
            yield n.val
            n = n.next

    def __repr__(self):
        return '[' + ', '.join(x for x in self) + ']'
```

```python
lst = LinkedList()
for i in range(10):
    lst.prepend(i)
lst
```

## 1.  Bidirectional links (Doubly-linked list) & Sentinel head



```python
class LinkedList:
    class Node:
        def __init__(self, val, prior=None, next=None):
            self.val = val
            self.prior = prior
            self.next  = next

    def __init__(self):
        self.count = 0

    def prepend(self, value):
        self.count += 1

    def append(self, value):
        self.count += 1

    def __getitem__(self, idx):
        pass

    def __len__(self):
        return self.count

    def __iter__(self):
        n = self.head.next
        while n is not self.head:
            yield n.val
            n = n.next

    def __repr__(self):
        return '[' + ', '.join(str(x) for x in self) + ']'
```

```python
lst = LinkedList()
for i in range(10):
    lst.prepend(i)
for i in range(10):
    lst.append(i)
lst
```

## 1.  Incorporating a "cursor"



```python
class LinkedList:
    class Node:
        def __init__(self, val, prior=None, next=None):
            self.val = val
            self.prior = prior
            self.next  = next

    def __init__(self):
        self.head = self.cursor = LinkedList.Node(None)
        self.head.prior = self.head.next = self.head
        self.count = 0

    def append(self, value):
        n = LinkedList.Node(value, prior=self.head.prior, next=self.head)
        n.prior.next = n.next.prior = n
        self.count += 1

    def cursor_set(self, idx):
        pass

    def cursor_insert(self, x):
        pass

    def cursor_delete(self):
        pass

    def __len__(self):
        return self.count

    def __iter__(self):
        n = self.head.next
        while n is not self.head:
            yield n.val
            n = n.next

    def __repr__(self):
        return '[' + ', '.join(str(x) for x in self) + ']'
```

```python
lst = LinkedList()
for i in range(10):
    lst.append(i)
lst
```

```python
lst.cursor_set(4)
for x in 'abcd':
    lst.cursor_insert(x)
lst
```

```python
lst.cursor_set(8)
for _ in range(4):
    lst.cursor_delete()
```
